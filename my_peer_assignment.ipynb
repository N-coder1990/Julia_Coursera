```julia
my_fun(x)=3*x^2+5
```

This is a test parabola function

```julia
input=Array{Int64}(-5:5)
output = Array{Int64}(undef,length(input))
for i = 1:length(input)
    output[i]=my_fun(input[i])
end
println(input)
println(output)

```

```julia
using Plots
gr()

plot(input,output,
title="function plot",

xlabel  = "x data",
ylabel  = "y data)",
marker  = (:diamond, 4),
line    = (:path, "black"),   
legend  = true,
grid    = true)
```
